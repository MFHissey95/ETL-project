```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
rds_connection_string = "postgres:delphicitb969@localhost:5432/education_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
csv_file = "education-statistics/EdStatsCountry.csv"
edstats_country_df = pd.read_csv(csv_file)
edstats_country_df.head()
```

```python
edstats_country = edstats_country_df[['Country Code', 'Short Name', 'Table Name', 'Long Name', '2-alpha code', 'Currency Unit', 'Region', 'Income Group']].copy()
edstats_country.head()
```

```python
edstats_country = edstats_country.rename(columns={"Country Code": "country_code",
                                                  "Short Name": "short_name",
                                                  "Table Name": "proper_name",
                                                  "Long Name":"long_name",
                                                  "2-alpha code": "two_alpha_code",
                                                  "Currency Unit": "currency_unit",
                                                  "Region": "region",
                                                  "Income Group": "income_group",})
edstats_country.head()
```

```python
csv_file = "education-statistics/EdStatsCountry-Series.csv"
edstats_country_series_df = pd.read_csv(csv_file)
edstats_country_series_df.head()
```

```python
edstats_country_series = edstats_country_series_df[['ID Unique','CountryCode','SeriesCode']].copy()
edstats_country_series.head()
```

```python
edstats_country_series = edstats_country_series.rename(columns={"ID Unique": "id_unique",
                                                                "CountryCode": "country_code",
                                                                "SeriesCode": "series_code"})
edstats_country_series.head()
```

```python
csv_file = "education-statistics/EdStatsSeries.csv"
edstats_series_df = pd.read_csv(csv_file)
edstats_series_df.head()
```

```python
edstats_series = edstats_series_df[['id_unique','Series Code', 'Topic', 'Indicator Name', 'Short definition']].copy()
edstats_series.head()
```

```python
edstats_series = edstats_series.rename(columns={"Series Code": "series_code",
                                                "Topic": "topic",
                                                "Indicator Name": "indicator_name",
                                                "Short definition":"short_definition"})
edstats_series.head()
```

```python
engine.table_names()
```

```python
edstats_country.to_sql(name='edstats_country', con=engine, if_exists='replace', index=True)
```

```python
edstats_country_series.to_sql(name='edstats_country_series', con=engine, if_exists='replace', index=True)
```

```python
edstats_series.to_sql(name='edstats_series', con=engine, if_exists='replace', index=True)
```
